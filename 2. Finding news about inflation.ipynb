```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from transformers import pipeline
import openpyxl
import pandas as pd
```

```python
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-xnli-multilingual-nli-2mil7")
```

```python
name = '2025-01-17'
news = pd.read_excel(f'Новости_исходники/{name}.xlsx')
news_lst = [text for text in news.text]
```

```python
%%time
label_hf = []
count = 0
for text in news_lst:
    candidate_labels = ["Инфляция и цены", "Не инфляция и не цены"]
    output = classifier(text, candidate_labels, multi_label=False)
    label_hf.append(output.get('labels')[0])
    count += 1
    print(f'Новость {count} обработана. Результат: {output.get("labels")[0]}')
```

```python
news['Label'] = label_hf
news.to_excel(f'Новости_исходники/{name}_Отсортированный.xlsx')
```
