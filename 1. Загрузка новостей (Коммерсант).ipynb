```python
from bs4 import BeautifulSoup
import pandas as pd
import requests, lxml, webbrowser, re, random, datetime
import time
import openpyxl
```

```python
# Основная ссылка https://www.kommersant.ru/archive/rubric/номер рубрики
# 2 - Политика
# 3 - Экономика
# 4 - Бизнес
# 40 - Финансы
# 41 - Потребительский рынок

category_lst = [3, 4, 41, 40] # [3, 4, 41, 2, 40]

pd.set_option('display.max_columns', None)

start = time.time()

# date_lst = ['09/02/2018', '23/03/2018', '27/04/2018', '15/06/2018','27/07/2018', '14/09/2018', '26/10/2018', '14/12/2018']
# date_lst = ['10/02/2023', '20/02/2023', '17/03/2023', '28/04/2023', '11/05/2023', '09/06/2023', '21/07/2023','31/07/2023', '15/08/2023',
#             '15/09/2023', '27/10/2023', '27/10/2023', '07/11/2023', '15/12/2023']
date_df = []
# for date in date_lst:
#     date_start = datetime.datetime.strptime(f'{date}','%d/%m/%Y').date()
#     # date_end = datetime.datetime.strptime('26/03/2018','%d/%m/%Y').date()
#     date_start = pd.to_datetime(date_start)
#     date_df.append(date_start)

date_start = datetime.datetime.strptime('03/12/2024','%d/%m/%Y').date()
date_end = datetime.datetime.strptime('17/01/2025','%d/%m/%Y').date()
# date_start = pd.to_datetime(date_start)
# date_df.append(date_start)
# print(date_df)
date_df = pd.date_range(date_start, date_end)
count = 0
print('Выгрузка данных началась!')
print()

news_for_the_month = pd.DataFrame(columns=['date', 'category', 'text', 'item', 'link']) # (columns=['date', 'category', 'text', 'item', 'link'])

for date in date_df:
    for category in category_lst:

        date_lst = []
        item_lst = []
        cat_lst = []
        text_lst = []
        link_lst = []
        news_df = pd.DataFrame(columns=['date', 'text', 'item', 'link']) #(columns=['date', 'text', 'item', 'link'])

        url = f'https://www.kommersant.ru/archive/rubric/{category}/day/{date.date()}'
        # print(url)
        req = requests.get(url)
        src = req.text
        soup = BeautifulSoup(src, 'lxml')
        # news = soup.find('h2', class_ = 'uho__name rubric_lenta__item_name')
        news_hub = soup.find_all('a', class_ = 'uho__link uho__link--overlay')

        for news in news_hub:
            news_ = news.text.strip()
            href = news.get('href')
            href_ = (f'https://www.kommersant.ru{href}')
            link_lst.append(href_)
            req_news = requests.get(href_)
            soup_news = BeautifulSoup(req_news.text, 'lxml')

            help_lst = []
    #        text_header = soup_news.find('h1', class_='doc_header__name js-search-mark') # Вытаскиваем заголовок
    #        text_subheader = soup_news.find('h2', class_='doc_header__subheader') # Вытаскиваем субзаголовок
            text_main = soup_news.find_all('p', class_='doc__text') # Основной текст статьи
            text_thought = soup_news.find_all('p', class_='doc__thought') # Комментарии
    #        help_lst.append(text_header.text.strip())
     #       help_lst.append(text_subheader.text.strip())

            for text in text_main:
                help_lst.append(text.text.strip())
            for thought in text_thought:
                help_lst.append(thought.text.strip())
            count += 1
            text_lst.append(' '.join(help_lst))
            item_lst.append(news_)
            date_lst.append(date.date())
            cat_lst.append(category)

        news_df['date'] = date_lst
        news_df['text'] = text_lst
        news_df['item'] = item_lst
        news_df['link'] = link_lst
        news_df['category'] = cat_lst

        # news_df.to_excel(f'Рубрики/Экономика/Новости за {date.date()}.xlsx', index=False)
        news_for_the_month = pd.concat([news_for_the_month, news_df], ignore_index=True)

        end = time.time()
        tim = end - start
        print(f'Категория: {category}. Новости за {date.date()} загружены. Скачено новостей: {count}. Затрачено времени с начала запуска программы: {round(tim // 60, 0)} м. {round(tim % 60, 0)} с.')

news_for_the_month.to_excel(f'Новости_исходники/{date.date()}.xlsx', index=False)


# news_df['date'] = date_lst
# news_df['text'] = text_lst
# news_df['item'] = item_lst
# news_df['link'] = link_lst

end = time.time()
tim = end - start

# news_df.to_excel(f'Новости за {date}.csv')

# print(f'На выгрузку {count} новостей {round(tim // 60, 0)} м. и {round(tim % 60, 0)}с.')

print()
print('Выгрузка завершена')
```
